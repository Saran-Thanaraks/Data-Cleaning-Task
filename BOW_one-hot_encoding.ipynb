```python
import pandas as pd
import numpy as np
data = pd.read_csv("Random_shopping_cart.csv")

data = data.replace({" ": None})
data = data.replace({np.nan: None})
data.head()
```

```python
unique_list = []
for col in data.columns:
    list = data[col].unique()
    for l in list:
        if l not in unique_list:
            unique_list.append(l)
    
print(unique_list)
```

```python
new_df = pd.DataFrame()
for stuff in unique_list:
    new_df[str(stuff).lstrip(" ")] = 0

new_df
```

```python
data_matrix = np.matrix(data)
for i in range(len(data_matrix)):
    row = np.array(data_matrix[i])[0]
    temp = {}
    for col in new_df.columns:
        temp[col] = [0]  
    for k in range(len(row)):
        for col in new_df.columns:
            if(col ==   str(row[k]).lstrip(" ")):
                temp[col] = [1]
    new_row = pd.DataFrame(temp)
    new_df = pd.concat([new_df, new_row], ignore_index=True)
new_df 
```

```python
index_list =[i for i in range(1,new_df.shape[0]+1)]
new_df.insert(0, 'TID', index_list)
new_df.set_index('TID', inplace=True)
new_df = new_df.drop(["None"], axis=1)
new_df
```

```python
new_df.to_csv("Cleaned_Random_shopping_cart.csv")
```
